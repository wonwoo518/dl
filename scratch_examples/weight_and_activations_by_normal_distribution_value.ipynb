```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
def sigmoid(x):
    return 1/(1+np.exp(-x))
```

```python
node_num = 100
def initRand(dist):
    
    input_x = np.random.randn(1000,node_num)
    hidden_layer_size = 5
    activations = {}

    for i in range(hidden_layer_size):
        if i != 0:
            input_x = activations[i-1]

        w = np.random.randn(node_num, node_num) * dist #정규분포 dist로 초기화
        a = np.dot(input_x,w)
        z = sigmoid(a)
        activations[i] = z
    showDist(w)
    showActivations(activations)
    

def showDist(dots):
    sns.distplot(dots.flatten())
    plt.show()
    
def showActivations(activations):
    for i, a in activations.items():
        plt.subplot(1, len(activations), i+1)
        plt.title(str(i+1) + '-layer')
        plt.hist(a.flatten(), 30, range=(0,1))

    plt.show()

initRand(1)
```

```python
initRand(0.01)
```

```python
initRand(1/np.sqrt(node_num))
```
